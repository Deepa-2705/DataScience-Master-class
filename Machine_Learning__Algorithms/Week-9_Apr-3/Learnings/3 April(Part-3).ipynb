### Logistic Regression for Multiclass Classification

```python
# make a prediction wtih a multinominal linguistic regression model 
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
# define dataset
x,y=make_classification(n_samples=1000,n_features=10,n_informative=5,n_redundant=5,n_classes=3,random_state=1)
```

### for multiclass problems, only 'newton-cg','sag','saga','lbfgs' handle multiclass loss
these are all the optimizers that helps to reach to the global minima 

```python
model=LogisticRegression(multi_class='ovr',solver='lbfgs')
```

```python
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
model.predict_proba(x_test)
```

```python
# confusion matrix,accuracy and recall precision
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))   # a 3*3 matrix because we have 3 categories in dependent variable 
# diagonal are accurate result and rest are incorrect result 
```
