Q1. Explain the concept of R-squared in linear regression models. How is it calculated, and what does it
represent?

Ans: R-squared is a goodness-of-fit measure for linear regression models. This statistic indicates the percentage of the variance in the dependent variable that the independent variables explain collectively.

R square
Formula

R^2 = 1 - SSR/SST

R^2 = coefficient of determination SSR = sum of squares of residuals SST = total sum of squares

An R-Squared value shows how well the model predicts the outcome of the dependent variable. R-Squared values range from 0 to 1. An R-Squared value of 0 means that the model explains or predicts 0% of the relationship between the dependent and independent variables.

Q2. Define adjusted R-squared and explain how it differs from the regular R-squared.

Ans: Adjusted R2 is a corrected goodness-of-fit (model accuracy) measure for linear models. It identifies the percentage of variance in the target field that is explained by the input or inputs. R2 tends to optimistically estimate the fit of the linear regression.
    
Adjusted R-squared is a modified version of R-squared that has been adjusted for the number of predictors in the model. The adjusted R-squared increases when the new term improves the model more than would be expected by chance.

Q4. What are RMSE, MSE, and MAE in the context of regression analysis? How are these metrics
calculated, and what do they represent?

Ans:
1) The Mean absolute error represents the average of the absolute difference between the actual and predicted values in the dataset. It measures the average of the residuals in the dataset.





2) Mean Squared Error represents the average of the squared difference between the original and predicted values in the data set. It measures the variance of the residuals.



3) Root Mean Squared Error is the square root of Mean Squared error. It measures the standard deviation of residuals.



Q5. Discuss the advantages and disadvantages of using RMSE, MSE, and MAE as evaluation metrics in
regression analysis.

Ans: 
* MSE

Advantage: The MSE is great for ensuring that our trained model has no outlier predictions with huge errors, since the MSE puts larger weight on theses errors due to the squaring part of the function.

Disadvantage: If our model makes a single very bad prediction, the squaring part of the function magnifies the error.

* MAE

Advantage: The beauty of the MAE is that its advantage directly covers the MSE disadvantage. Since we are taking the absolute value, all of the errors will be weighted on the same linear scale. Thus, unlike the MSE, we wont be putting too much weight on our outliers and our loss function provides a generic and even measure of how well our model is performing.

Disadvantage: If we do in fact care about the outlier predictions of our model, then the MAE wont be as effective. The large errors coming from the outliers end up being weighted the exact same as lower errors. This might results in our model being great most of the time, but making a few very poor predictions every so-often.

* RMSE

Advantage: Like MSE, RMSE is dependent on the scale of the data. It increases in magnitude if the scale of the error increases. 

Disadvantage: One major drawback of RMSE is its sensitivity to outliers and the outliers have to be removed for it to function properly. RMSE increases with an increase in the size of the test sample.

Q6. Explain the concept of Lasso regularization. How does it differ from Ridge regularization, and when is
it more appropriate to use?

Ans: Lasso regression, commonly referred to as L1 regularization, is a method for stopping overfitting in linear regression models by including a penalty term in the cost function. In contrast to Ridge regression, it adds the total of the absolute values of the coefficients rather than the sum of the squared coefficients.

while lasso regression takes the magnitude of the coefficients, ridge regression takes the square.

The lasso procedure encourages simple, sparse models (i.e. models with fewer parameters). This particular type of regression is well-suited for models showing high levels of multicollinearity or when you want to automate certain parts of model selection, like variable selection/parameter elimination.

Q9. You are comparing the performance of two regression models using different evaluation metrics.
Model A has an RMSE of 10, while Model B has an MAE of 8. Which model would you choose as the better
performer, and why? Are there any limitations to your choice of metric?

Ans: The choice between Model A and Model B as the better performer depends on the specific context and priorities of your problem, as well as the characteristics of your data. Let's break down the differences between RMSE (Root Mean Square Error) and MAE (Mean Absolute Error) and consider their implications:

1) RMSE (Root Mean Square Error):

RMSE gives more weight to larger errors compared to smaller errors due to the squaring operation.
It is sensitive to outliers and tends to penalize them more heavily because of the squared term.
RMSE is commonly used when the errors are assumed to have a normal distribution or when larger errors should be penalized more.

2) MAE (Mean Absolute Error):

MAE treats all errors equally because it doesn't involve squaring the errors.
It is less sensitive to outliers compared to RMSE, as it doesn't exaggerate their impact.
MAE is useful when you want to measure the average magnitude of errors without giving undue weight to large errors.
To choose between Model A and Model B, you should consider the following:

If your problem is relatively robust to outliers or you have reasons to believe that outliers are not a significant concern, and you want a metric that provides a straightforward average of prediction errors, then MAE (Model B) might be a better choice. In this case, Model B would be preferred because it has a lower MAE (8).

If your problem is sensitive to large errors or you want to emphasize the reduction of large errors in your predictions, then RMSE (Model A) might be more appropriate. RMSE is influenced more by outliers, and a lower RMSE (10) suggests that Model A may have fewer large errors.

Consider the domain-specific implications. Sometimes, certain errors may have more severe consequences than others, and this should be taken into account when choosing a metric.

Keep in mind that the choice of metric should align with the specific goals and requirements of your project. There is no universally "better" metric; it depends on the nature of the problem and what you value most in your predictions.

In summary, Model B (with a lower MAE of 8) is better if you want to minimize the average magnitude of errors, especially when outliers are not a significant concern. However, if you want to emphasize the reduction of larger errors or if your problem is sensitive to outliers, Model A (with a lower RMSE of 10) may be preferred. It's essential to understand the context and priorities of your problem to make an informed choice.

Limitations to consider when choosing a metric:

The choice of metric may not consider all aspects of model performance, so it's often a good practice to evaluate models using multiple metrics and consider the overall picture.
The choice of metric can be influenced by the distribution of errors, so it's essential to examine the distribution of errors and understand its characteristics.
Metrics alone may not capture all the nuances of model performance. Qualitative analysis, domain knowledge, and business objectives should also be taken into account.

Q10. You are comparing the performance of two regularized linear models using different types of
regularization. Model A uses Ridge regularization with a regularization parameter of 0.1, while Model B
uses Lasso regularization with a regularization parameter of 0.5. Which model would you choose as the
better performer, and why? Are there any trade-offs or limitations to your choice of regularization
method?

Ans: The choice between Ridge regularization (L2 regularization) and Lasso regularization (L1 regularization) depends on your specific problem, dataset, and goals. Let's compare the two regularization methods and consider their implications:

* Ridge Regularization (L2):

Ridge regularization adds a penalty term to the linear regression loss function based on the sum of the squares of the model's coefficients.

It encourages the model to keep all features, but it reduces the magnitude of the coefficients, effectively shrinking them toward zero.

Ridge regularization is generally suitable when you have many features, and you believe that most of them are relevant, but you want to prevent multicollinearity or control the size of the coefficients.

* Lasso Regularization (L1):

Lasso regularization adds a penalty term to the linear regression loss function based on the absolute values of the model's coefficients.

It encourages sparsity by driving some of the coefficients to exactly zero, effectively performing feature selection.
Lasso regularization is suitable when you suspect that only a subset of features is relevant, and you want the model to automatically select the most important ones.

Now, to choose between Model A (Ridge) and Model B (Lasso) with their respective regularization parameters:

If your dataset has many features, and you believe that most of them are potentially relevant to the prediction, Ridge regularization (Model A) might be more appropriate with a regularization parameter of 0.1. This value suggests moderate regularization, which would help control the magnitudes of coefficients without driving any of them to zero.

If you suspect that your dataset has many features, but only a subset of them is truly relevant, and you want automatic feature selection, Lasso regularization (Model B) with a regularization parameter of 0.5 could be a better choice. The higher regularization strength (0.5) in Lasso is more likely to drive some coefficients to exactly zero, effectively performing feature selection.

* Trade-offs and limitations:

Ridge tends to be more robust when there is multicollinearity among the features, as it does not force coefficients to be exactly zero. Lasso may struggle with multicollinearity and select one of the correlated features while driving others to zero.

Lasso is better suited for feature selection, but it may not be ideal when you genuinely believe that all features are relevant. It can lead to an overly sparse model if used indiscriminately.

The choice between Ridge and Lasso depends on the specific problem and the importance of feature selection. Sometimes, a combination of both (Elastic Net) is used to benefit from the strengths of both regularization techniques.

-> In summary, the choice between Model A (Ridge) and Model B (Lasso) depends on whether you prefer to retain all features with controlled coefficients (Ridge) or automatically select a subset of relevant features (Lasso). The specific context of your problem, your understanding of the data, and your goals should guide your choice of regularization method and strength.
