```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.random.rand(50)
```

```python
y=np.random.rand(50)
```

```python
x
```

```python
y
```

### Scatter plot

```python
 plt.scatter(x,y)
```

```python
# Make more informative scatter plot
# Alpha will control the intensity of the colour  range is : 0 - 1
# Marker : represent the coordinates with a symbol
plt.figure(figsize=(8,2))   # Place this figsize before the any plot (on top of every line)
plt.scatter(x,y,c='#2AEE98',alpha=1,marker='x')   # c = 'r' red colour
plt.xlabel("Hour studied")
plt.ylabel("Marks obtained")
plt.title("This is Hour studied vs Marks obtained")
plt.grid()  # place the boxes
```

```python
x=np.linspace(1,10,100)   #Generate random data having value from 1 to 10 and size is 10
y=np.sin(x)
```

```python
x
```

```python
y
```

### Line plot

```python
plt.plot(x,y,c='r')  # It plot dot and connect them
```

```python
plt.scatter(x,y)  #It will just plot the dots
```

```python
plt.figure(figsize=(12,2)) #Control the size of the figure on the window 
plt.plot(x,y,'--r')   # It will plot not the continuous line but the broken line
plt.xlabel("Hour studied")
plt.ylabel("Marks obtained")
plt.title("This is sine wave graph")
plt.grid()  # place the boxes
```

### Barplot 

```python
x=["a","b","c","d","e"]   # categorical data 
```

```python
y=np.random.rand(5)      #Integer data
```

```python
x
```

```python
y
```

```python
plt.figure(figsize=(6,2)) #Control the size of the figure on the window 
plt.bar(x,y,color='pink')
plt.xlabel("Hour studied")
plt.ylabel("Marks obtained")
plt.title("This is bar plot")
plt.grid()  # place the boxes
```

```python
# Flip the bar
plt.figure(figsize=(6,4)) #Control the size of the figure on the window 
plt.barh(x,y,color='purple')
plt.xlabel("Hour studied")
plt.ylabel("Marks obtained")
plt.title("This is bar plot")
plt.grid()  # place the boxes
```

### 3-D plot

```python
x=np.random.rand(50)
```

```python
y=np.random.rand(50)
```

```python
z=np.random.rand(50)
```

```python
# Create a figure first
fig=plt.figure()
# inside fig define plot this figure into 3-D graph
vis=fig.add_subplot(projection='3d')
vis.scatter(x,y,z,c='r',alpha=0.9)
plt.xlabel("x")
plt.ylabel("y")
plt.title("3-D plot")
plt.show()
```
