### Feature Scaling
1) Standardization
2) Normalization- min max scaler
3) Unit vector

```python
# 1) Standardization
import seaborn as sns 
```

```python
df=sns.load_dataset('tips')
```

```python
df.head()
```

```python
# Apply standardization on total_bill column
```

```python
# calculate mean and sd
import numpy as np
mean=np.mean(df['total_bill'])
sd=np.std(df['total_bill'])   
print(mean,sd)
```

```python
# apply z-score formula now 
standarized_data=[]
for i in list(df['total_bill']):
    z_score=(i-mean)/sd
    standarized_data.append(z_score)
```

```python
standarized_data
```

```python
sns.histplot(df['total_bill'])
```

```python
sns.histplot(standarized_data)    # The distribution is not change but the dimension is reduced to smaller value
```

```python
# sklearn.preprocessing.StandardScaler : This library of scikit learn will apply standardization on every datapoints
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# standardscaler will do z-score calculation for every datapoints
#initialize standardscaler
scaler=StandardScaler()  # scaler is a object
scaler
```

```python
scaler.fit(df[['total_bill']])  # fit: it compute mean and sd of the datasets
```

```python
scaler.transform(df[['total_bill']])  # transform will apply z-score formula and then scale it down
```

```python
# shorter method
scaler.fit_transform(df[['total_bill']])
```

```python
# convert into dataframe
import pandas as pd 
pd.DataFrame(scaler.fit_transform(df[['total_bill']]),columns=['total_bill'])
```

```python
# We can also standardized 2 or more feature at the same time 
import pandas as pd 
pd.DataFrame(scaler.fit_transform(df[['total_bill','tip']]),columns=['total_bill','tip'])
```

```python
# What the need of that object ?
# When we get new datasets then we have to perform feature engineering again on that data. So using the object we can easily tranform it.
scaler.transform([[23,13]])    # 23 and 13 are new datasets 
```

```python
#2) Normalization- min max scaler
df=sns.load_dataset('taxis')
df.head()
```

```python
# Min max scalar : it bring down the numeric datasets to 0 to 1
from sklearn.preprocessing import MinMaxScaler
```

```python
# Initialize an object 
min_max=MinMaxScaler()
```

```python
min_max.fit_transform(df[['distance','fare','tip']])  
```

```python
min_max.fit(df[['distance','fare','tip']])   # fit: it compute xmin and xmax and other values of the datasets
```

```python
min_max.transform(df[['distance','fare','tip']])   # transform will apply z-score formula and then scale it down
```

```python
# we can also compute new datasets using the min_max object
min_max.transform([[1.60,27.0,2.15]])
```

```python
# 3) Unit vector
from sklearn.preprocessing import normalize
```

```python
normalize(df[['distance','fare','tip']])
```

```python
# Convert it into dataframe
pd.DataFrame(normalize(df[['distance','fare','tip']]))  # reduces large datasets to its unit value 
```
